```python
from random import randint
from matplotlib import pyplot as plt
```

```python
ground_truth = {
    1: 15,
    2: 25,
    3: 35,
    4: 45,
    5: 55,
    6: 65,
    7: 75,
    8: 85,
    9: [25, 45, 65, 85]
} 
```

```python
start_batch = 8
end_batch = 8
```

```python
for i_1 in range(1,9):
    gr = ground_truth[i_1]
    for i_2 in range(2,3):

        path = 'output/events_dice_0{}-{}/output_new.txt'.format(i_1, i_2)
        results = pd.read_csv('{}'.format(path), sep=" ", header=None)
        results.columns = ['index', "ts", "te", "loss", "depth", "wx", "wy", "wz"]
        
        #add sequence indication
        results['sequence'] = '{}-{}'.format(i_1, i_2)

        #add time difference
        results['td'] = results['te'] - results['ts']

        # recalculate time starting with 0
        timestamp = [0]
        for i in range(len(results)):
            timestamp.append(results['td'][i]+timestamp[i])
        results['timestamp'] = timestamp[0:-1]

        # append to overal results table
        overal = overal.append(results[['sequence', 'timestamp', 'depth']])

        time = results['timestamp'][start_batch:-end_batch]
        depth = results['depth'][start_batch:-end_batch]*100

        plt.plot(time, depth, label = '{}-{}'.format(i_1, i_2), linestyle=":")

        average_depth = depth.mean()

        plt.plot(time, [average_depth]*len(depth), label = 'average {}-{}'.format(i_1, i_2), color=plt.gca().lines[-1].get_color())
        plt.plot(time, [gr]*len(depth), label = 'ground truth {}-{}'.format(i_1, i_2), linestyle="--", color=plt.gca().lines[-1].get_color())

        plt.legend()
        plt.show()
```
